```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
node_data = pd.read_csv('data_train_edge.csv')
predict_data = pd.read_csv('predict.csv')
ans_data = pd.read_csv('ans500_ground_truth.csv')
```

```python
G = nx.from_pandas_edgelist(node_data, "node1", "node2")
```

```python
adamic_adar_index = []
preferential_attachment = []
jaccard_coefficient = []
```

```python
for i in range(len(predict_data)):
    try:
        preds = nx.adamic_adar_index(G, [(predict_data['node1'][i], predict_data['node2'][i])])
        for u, v, p in preds:
            adamic_adar_index.append(p)
    except:
        adamic_adar_index.append(0)
        
for i in range(len(predict_data)):
    try:
        preds = nx.preferential_attachment(G, [(predict_data['node1'][i], predict_data['node2'][i])])
        for u, v, p in preds:
            preferential_attachment.append(p)
    except:
        preferential_attachment.append(0)
        
for i in range(len(predict_data)):
    try:
        preds = nx.jaccard_coefficient(G, [(predict_data['node1'][i], predict_data['node2'][i])])
        for u, v, p in preds:
            jaccard_coefficient.append(p)
    except:
        jaccard_coefficient.append(0)
```

```python
trainX_adamic = pd.DataFrame()
trainX_prefer = pd.DataFrame()
trainX_jaccar = pd.DataFrame()
# trainX = pd.DataFrame()
```

```python
trainX_adamic['train'] = adamic_adar_index[:500]
trainX_prefer['train'] = preferential_attachment[:500]
trainX_jaccar['train'] = jaccard_coefficient[:500]
predict_all = pd.DataFrame()
predict_all['adamic'] = adamic_adar_index
predict_all['prefer'] = preferential_attachment
predict_all['jaccar'] = jaccard_coefficient
print(predict_all.shape)
predict_data['adamic'] = adamic_adar_index
predict_data['prefer'] = preferential_attachment
#predict_data['jaccar'] = jaccard_coefficient
print(predict_data.shape)
```
